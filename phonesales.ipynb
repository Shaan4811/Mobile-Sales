```python
sns.pairplot(data,hue='Brand');
```

```python
fig, ax = plt.subplots(figsize=(15,3))
ax=sns.countplot(x="Brand", data=data)
plt.title('Most Used Brand');
```

```python
# it is clear that samsung users are high
```

```python
plt.figure(figsize=(10,6))
sns.boxplot(x='Brand',y='Selling Price',data=data,palette='bright')
plt.xticks(rotation=90)
plt.title('Price range of different brands')
plt.show()
```

```python
data["Price Range"] = pd.cut(data["Selling Price"], bins=[0, 15000, 40000, float("inf")], labels=["Low Range(below 15k)", "Mid Range(15k-40k)", "Premium Range(above 40k)"])

plt.figure(figsize=(10, 6))
sns.countplot(x="Price Range",data=data,palette="flare")
plt.title("Count of Phones in Different Price Ranges")
plt.xlabel("Price Range")
plt.ylabel("Count")
plt.show()
```

```python
# Most Smartphones Belong to Low Range Category (0-15000 Rs)
```

```python
common_colors = data["Color"].value_counts().head(10)
plt.figure(figsize=(8, 8))
plt.pie(common_colors,labels=common_colors.index,autopct='%1.1f%%',startangle=140,colors=sns.color_palette("muted"),explode=[0.2,0,0,0,0,0,0,0,0,0],shadow=True)
plt.title("Most Common Colors Offered")
plt.show()
```

```python
# Most commonly used color is black
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x="Brand",y="Rating",data=data,palette="bright")
plt.xticks(rotation=90)
plt.title('Rating according to Brands')
plt.show()
```

```python
sns.countplot(x="Memory",data=data,palette="flare")
plt.xticks(rotation=90)
plt.title('Distribution of RAM')
plt.show()
```

```python
sns.countplot(x="Storage",data=data,color="r")
plt.xticks(rotation=90)
plt.title('Distribution of Storage')
plt.show()
```
